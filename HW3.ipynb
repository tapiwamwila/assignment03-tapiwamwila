```python
z = velo[:, 0]  # Depth data
v = velo[:, 1]  # Measured velocity data

#  window sizes in meters (number of points in the window)
window_sizes = [3, 10, 50]

#  the standard deviation for Gaussian kernel (controls the weighting)
std_dev_factor = 0.5  # You can adjust this to control how sharply the weights decay


plt.figure(figsize=(10, 6))
plt.plot(z, v, 'ko', markersize=4, label='Measured Data')  # Original data

# Function to create a Gaussian kernel
def gaussian_window(window_size, std_dev):
    # Generate an array of positions around the center (zero-centered)
    positions = np.arange(window_size) - window_size // 2
    # Calculate the Gaussian weights using the standard normal distribution
    weights = norm.pdf(positions, scale=std_dev * window_size)
    return weights / np.sum(weights)  # Normalize the weights

# Function to apply weighted moving average using the Gaussian kernel
def weighted_moving_average(data, window_size, std_dev):
    # Generate a Gaussian window
    gauss_window = gaussian_window(window_size, std_dev)
    # Apply the Gaussian weighted window using convolution
    smoothed_data = convolve1d(data, gauss_window, mode='reflect')
    return smoothed_data

# Looping  over each window size to calculate the weighted moving average and plot
for window in window_sizes:
    # Apply the weighted moving average
    v_smoothed = weighted_moving_average(v, window, std_dev_factor)
    
    # lets plot the weighted moving average
    plt.plot(z, v_smoothed, label=f'Weighted Moving Avg (window={window} meters)', linewidth=2)


plt.xlabel('Depth (m)')
plt.ylabel('Velocity (m/yr)')
plt.title('Velocity as a function of Depth with Weighted Moving Window Average')
plt.legend()
plt.grid(True)
plt.show()
```

```python

z = velo[:, 0]  # Depth data
v = velo[:, 1]  # Measured velocity data

# Define the standard deviation factor for Gaussian kernel (this can be adjusted)
std_dev_factor = 0.5

# Function to create a Gaussian kernel
def gaussian_window(window_size, std_dev):
    positions = np.arange(window_size) - window_size // 2
    weights = norm.pdf(positions, scale=std_dev * window_size)
    return weights / np.sum(weights)

# Function to apply the weighted moving average
def weighted_moving_average(data, window_size, std_dev):
    gauss_window = gaussian_window(window_size, std_dev)
    smoothed_data = convolve1d(data, gauss_window, mode='reflect')
    return smoothed_data

# Function to calculate RMSE
def calculate_rmse(v_true, v_pred):
    return np.sqrt(np.mean((v_true - v_pred) ** 2))

# Define a range of window sizes to evaluate
window_range = np.arange(3, 101, 2)  # Window sizes from 3 to 100, step 2
rmse_values = []

# Loop over each window size and compute RMSE
for window in window_range:
    v_smoothed = weighted_moving_average(v, window, std_dev_factor)
    rmse = calculate_rmse(v, v_smoothed)
    rmse_values.append(rmse)

# Find the window size that minimizes RMSE
optimal_window = window_range[np.argmin(rmse_values)]
optimal_rmse = min(rmse_values)

#we now plot the RMSE vs window size
plt.figure(figsize=(10, 6))
plt.plot(window_range, rmse_values, label='RMSE', color='blue', linewidth=2)
plt.axvline(optimal_window, color='red', linestyle='--', label=f'Optimal Window = {optimal_window} (RMSE={optimal_rmse:.4f})')
plt.xlabel('Window Size (meters)')
plt.ylabel('RMSE')
plt.title('RMSE vs Window Size for Weighted Moving Average')
plt.legend()
plt.grid(True)
plt.show()

# Print the optimum window size and corresponding RMSE
print(f"Optimal window size: {optimal_window} meters")
print(f"Minimum RMSE: {optimal_rmse:.4f}")
```

```python
# Constants
rho = 917  # Density of ice (kg/m^3)
g = 9.81  # Gravity (m/s^2)
theta = np.radians(10)  # Surface slope (radians)

# Measured data (depth, velocity)
z = velo[:, 0]  # Depth data (in meters)
v_measured = velo[:, 1]  # Measured velocity data (in m/yr)

# Measured surface velocity (at z = 0)
u_surf = v_measured[0]

# Grid search parameter ranges
A_values = np.linspace(1e-25, 1e-16, 50)  # A values to try
n_values = np.linspace(1, 5, 50)  # n values to try

# Function to calculate velocity from A and n
def calculate_velocity(A, n, z, u_surf):
    return u_surf - A * (rho * g * np.sin(theta))**n * z**(n + 1)

# Function to calculate RMSE
def calculate_rmse(v_true, v_pred):
    return np.sqrt(np.mean((v_true - v_pred) ** 2))

# Initialize variables to store results
best_rmse = float('inf')
best_A, best_n = None, None
rmse_grid = np.zeros((len(A_values), len(n_values)))

# Grid search over A and n
for i, A in enumerate(A_values):
    for j, n in enumerate(n_values):
        # Predicted velocities for the current A and n
        v_predicted = calculate_velocity(A, n, z, u_surf)
        
        # Calculate RMSE
        rmse = calculate_rmse(v_measured, v_predicted)
        rmse_grid[i, j] = rmse
        
        # Update best parameters if this RMSE is smaller
        if rmse < best_rmse:
            best_rmse = rmse
            best_A, best_n = A, n

# Print the best A and n values
print(f"Optimal A: {best_A:.4e}")
print(f"Optimal n: {best_n:.4f}")
print(f"Minimum RMSE: {best_rmse:.4f}")

# Plot the RMSE as a function of A and n
plt.figure(figsize=(8, 6))
plt.imshow(rmse_grid, extent=[n_values.min(), n_values.max(), A_values.min(), A_values.max()],
           aspect='auto', origin='lower', cmap='viridis')
plt.colorbar(label='RMSE')
plt.xlabel('n (flow law exponent)')
plt.ylabel('A (flow law parameter)')
plt.title('RMSE for A and n in Grid Search')
plt.show()
```

```python
# Constants
rho = 917  # Density of ice (kg/m^3)
g = 9.81  # Gravity (m/s^2)
theta = 10 * np.pi / 180  # Surface slope in radians

# Measured data (replace with actual data)
z = np.array([0, 10, 20, 30, 40, 50])  # Depth (m)
v_measured = np.array([30, 28, 26, 24, 22, 20])  # Velocity (m/yr) at corresponding depths

# Surface velocity (at z = 0)
u_surf = v_measured[0]

# Function to calculate velocity based on A and depth
def calculate_velocity(A, n, z, u_surf):
    return u_surf - A * (rho * g * np.sin(theta))**n * z**(n+1)

# Objective function to minimize (RMSE)
def objective_function(A, z, v_measured, u_surf, n=3):
    v_predicted = calculate_velocity(A, n, z, u_surf)
    rmse = np.sqrt(np.mean((v_predicted - v_measured)**2))
    return rmse

# Number of Monte-Carlo simulations
num_simulations = 1000

# Arrays to store optimized A values and corresponding RMSE
A_values = []
rmse_values = []

# Perform 1000 random samplings of 90% of the data
for _ in range(num_simulations):
    # Randomly sample 90% of the data
    indices = random.sample(range(len(z)), int(0.9 * len(z)))
    z_sample = z[indices]
    v_sample = v_measured[indices]
    
    # Initial guess for A
    initial_guess = [1e-20]
    
    # Optimize A using the gradient search method (minimize)
    result = minimize(objective_function, initial_guess, args=(z_sample, v_sample, u_surf), method='Nelder-Mead')
    
    # Store the optimized A and RMSE values
    A_opt = result.x[0]
    A_values.append(A_opt)
    rmse_values.append(result.fun)

# Calculate mean and standard deviation of A values
A_mean = np.mean(A_values)
A_std = np.std(A_values)

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.hist(A_values, bins=30, density=True, color='b', alpha=0.7)
plt.xlabel('Optimized A')
plt.ylabel('Relative Density')
plt.title('Distribution of A values')

# Plot the mean and standard deviation of A with vertical error bars
plt.errorbar(A_mean, 0.05, xerr=A_std, fmt='o', color='r', label=f'Mean A: {A_mean:.2e} ± {A_std:.2e}', capsize=5)
plt.legend()

# Plot histogram for RMSE values
plt.subplot(1, 2, 2)
plt.hist(rmse_values, bins=30, density=True, color='r', alpha=0.7)
plt.xlabel('RMSE')
plt.ylabel('Relative Density')
plt.title('Distribution of RMSE values')

plt.tight_layout()
plt.show()

print(f"Mean value of A: {A_mean:.4e}")
print(f"Standard deviation of A: {A_std:.4e}")
```
